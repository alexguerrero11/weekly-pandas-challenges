----
### 3. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline.

```python
data.columns #listing all columns in dataset
```

```python
discipline = data[['Team','Yellow Cards','Red Cards']] #assign certain column to dataset discipline
discipline
```

----
### 4. Sort the teams by Red Cards, then to Yellow Cards.

```python
discipline.sort_values(by=['Red Cards', 'Yellow Cards'])
```

----
### 5. Calculate the mean Yellow Cards given per Team.

```python
yellow_card_mean = discipline['Yellow Cards'].mean()
print(f'The mean of yellow card per team is',yellow_card_mean)
```
