### 5. Turn the item price into a float

```python
chipotle['item_price'] = chipotle['item_price'].str.replace('$','') #replacing '$' with blank
chipotle['item_price'] = chipotle['item_price'].astype(float) #assigning type as float
```

```python
chipotle.dtypes #confirming type
```

```python
chipotle.head()
```

----
### 6. How much was the revenue for the period in the dataset?

```python
total_revenue = chipotle['item_price'].sum()
print(f"The total revenue for the time period of the dataset is ${total_revenue}.")
```

----
### 7. How many orders were made in the period?

```python
#get count of unique order id
total_count_order = len(chipotle['order_id'].unique()) #got array of unique order_id
print(f'During this period there were {total_count_order} total orders made.')
```

----
### 8. What is the average revenue amount per order?

```python
avg_per_order = round(total_revenue/total_count_order,2) #rounding the avg into 2 decimals
print(f'The average revenue amount per order is ${avg_per_order}.')
```
